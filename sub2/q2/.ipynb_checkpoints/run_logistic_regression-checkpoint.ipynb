Question 2.2 c) As training progresses, the ce for validation first decreases then increases, whereas it decreases contineously for training data. At every stage, the ce for validation is higher than for training. These all make sence because initially the model is not trained enough so it is unable to capture important features in our dataset - fits both the training and validation bad. As we do more training, our model will fit the training dataset better, and its performance on the validation dataset will increase. However, if we have trained our model too much, then it becomes too specific to the training dataset, overfitting it. It fails to generalize, so validation ce increases. If we run multiple times, the results will not change because of the way we initialize our weights with all zeros. If they do change, we can do a number of re-runs and average the ce/errors to find the best parameter settings.

A lambda that is too large makes the model underfits, failing to capture important features from the training dataset. The model will not perform well on both training and validation datasets.
